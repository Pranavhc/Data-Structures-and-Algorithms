### Singly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class SynglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self, data):
        self.length += 1
        if self.head == None:
            self.head = self.tail = Node(data)
        else:
            self.tail.next = Node(data)
            self.tail = self.tail.next

    def prepend(self, data):
        newNode = Node(data)
        self.length += 1

        if self.head == None:
            self.head = self.tail = newNode
        else:
            newNode.next = self.head
            self.head = newNode

    def insertAt(self, data, idx):
        if idx == 0:
            self.prepend(data)
            return
        elif idx == self.length:
            self.append(data)
            return
        elif idx > self.length:
            return

        newNode = Node(data)
        prevNode = self.getPrevNodeByIdx(idx)

        newNode.next = prevNode.next
        prevNode.next = newNode
        self.length += 1

    def get(self, idx):
        if idx == 0:
            return self.head.data
        return self.getPrevNodeByIdx(idx).next.data

    def remove(self, data):
        if self.head.data == data:
            out = self.head.data
            self.head = self.head.next
            self.length -= 1
            return out

        prev = self.getPrevNodeByData(data)
        out = prev.next.data

        if prev == None or prev.next == None:
            raise Exception("You can't delete what isn't there mate!")

        prev.next = prev.next.next
        self.length -= 1

        return out

    def removeAt(self, idx):
        if idx == 0:
            out = self.head.data
            self.head = self.head.next
            self.length -= 1
            return out

        prev = self.getPrevNodeByIdx(idx)
        out = prev.next.data

        if prev == None or prev.next == None:
            raise Exception("You can't delete what isn't there mate!")

        prev.next = prev.next.next
        self.length -= 1

        return out

    def reverse(self):
        curr = self.head
        prev = None
        nextNode = None

        while curr != None:
            nextNode = curr.next    
            curr.next = prev
            prev = curr
            curr = nextNode

        self.head = prev

    def getPrevNodeByIdx(self, idx):
        curr = self.head
        for i in range(idx - 1):
            curr = curr.next
        return curr

    def getPrevNodeByData(self, data):
        curr = self.head
        for i in range(self.length - 1):
            if curr.next.data == data:
                break
            curr = curr.next
        return curr

    def printList(self):
        curr = self.head
        while curr != None:
            print(curr.data, end=" ")
            curr = curr.next
        print()
```
