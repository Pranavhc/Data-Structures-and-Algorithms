```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def at(self, idx):
        if idx > self.length:
            return None
        return self.__getNodeAt(idx).data;

    def append(self, data):
        node = Node(data)
        self.length += 1
        if self.head == None:
            self.head = node
            self.tail = self.head
        else:
            self.tail.next = node
            node.prev = self.tail

    def prepend(self, data):
        node = Node(data)
        self.length += 1
        if self.head == None:
            self.head = node
            self.tail = self.head
        else:
            node.next = self.head
            self.head.prev = node
            self.head = node

    def insertAt(self, data, idx):
        if idx == 0:
            self.prepend(data)
            return
        elif idx == self.length:
            self.append(data)
            return

        node = self.__getNodeAt(idx)

        if node == None:
            return

        newNode = Node(data)

        newNode.prev = node.prev
        newNode.next = node
        node.prev.next = newNode
        node.prev = newNode

        self.length += 1

    def remove(self, data):
        return self.__removedNode(self.__getNode(data))

    def removeAt(self, idx):
        return self.__removedNode(self.__getNodeAt(idx))

    def __removedNode(self, node):
        if node == None or self.head == None:
            return
        out = node.data

        if self.head == node:
            self.head = self.head.next
            self.head.prev = None
        elif self.tail == node:
            self.tail = self.tail.prev
            self.tail.next = None
        else:
            node.prev.next = node.next
            node.next.prev = node.prev
            self.length -= 1

        self.length -= 1
        return out

    def __getNodeAt(self, idx):
        if idx > self.length:
            return
        curr = self.head
        for i in range(idx):
            curr = curr.next
        return curr

    def __getNode(self, data):
        curr = self.head
        while curr:
            if curr.data == data:
                break
            curr = curr.next
        return curr

    def printList(self):
        curr = self.head
        while curr:
            print(curr.data, end=" ")
            curr = curr.next
        print()
```
